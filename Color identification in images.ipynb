# Task1 : Color Identification in Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('C:/Users/hp/Documents/PRIYANKA FOLDER/grass.jpg')
```

```python
plt.figure(figsize=(20,8))
plt.imshow(image)
```

```python
grid_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(grid_RGB)
```

```python
grid_HSV = cv2.cvtColor(grid_RGB, cv2.COLOR_RGB2HSV)
```

```python
lower = np.array([35,150,50])
upper = np.array([75,255,255])
```

```python
mask = cv2.inRange(grid_HSV, lower, upper)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

```python
res = cv2.bitwise_and(grid_RGB, grid_RGB, mask=mask)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(res)
```
